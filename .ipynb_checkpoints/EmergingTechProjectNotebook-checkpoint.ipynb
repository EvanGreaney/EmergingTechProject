# Emerging Technologies Project 2020

## Project Overview

For this project I was tasked with producing a model that accurately predicts wind turbine power output from wind speed values from a dataset powerproduction I was given, The idea is to use machine learning to produce an accurate prediction of the dataset.

```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
```

To start of the project, I begin by reading in the dataset from the CSV using pandas,then displaying the dataset to show that the data is being read in successfully. From here we can see that all five hubdred rows of data are successfuly being read in  and it is showing the first five rows and the last five rows

```python
# reads dataset from CSV file given as part of project
dataset = pd.read_csv("https://raw.githubusercontent.com/ianmcloughlin/2020A-machstat-project/master/dataset/powerproduction.csv")
dataset
```

Once the data is successfully read , the next logical step was to visualise to try and visualise the data to understand the to see how much power was being generated based on the speed of the wind. 

For the plotting of the graph matplotlib.pyplot library to plot to a scatter graph

Upon inspection of the graph, I can see that the data when plotted is sigmoid in shape and looks like a sigmoid function on the graph similar to a logistic curve/function

```python
# plotting of data from the dataset to a scatter plot graph
plt.figure(figsize=(20,10))
plt.title('Wind Turbine Power vs Wind Speed', size = 25)
plt.plot(dataset['speed'], dataset['power'],'bx')
plt.xlabel('Speed', size = 25)
plt.ylabel('Power', size = 25)
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.show
```

I then decided to store the data set in inside a a 2 dimensional data structure called fulldataset as to store it in a more SQL style table.

```python
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html
#with outliers
fulldataset = pd.DataFrame(dataset)
fulldataset
```

I then used keras to model my data, I decided to instantiate with the "Functional API" where you start at input and end with output and chain layers to specify the models forward pass between the input and output

For my first layer I set it the input with a node of 1 and then I created a hidden layer with 20 nodes and then nine more hidden layers with 80 nodes and all of them with an activation of sigmoid as the dataset is sigmoid in hsape when plotted.

```python
# https://towardsdatascience.com/multi-layer-neural-networks-with-sigmoid-function-deep-learning-for-rookies-2-bf464f09eb7f
# https://keras.io/api/models/model/
inputs = keras.Input(shape=(1,))
x1 = layers.Dense(20, activation="sigmoid")(inputs)
x2 = layers.Dense(80, activation="sigmoid")(x1)
x3 = layers.Dense(80, activation="sigmoid")(x2)
x4 = layers.Dense(80, activation="sigmoid")(x3)
x5 = layers.Dense(80, activation="sigmoid")(x4)
x6 = layers.Dense(80, activation="sigmoid")(x5)
x7 = layers.Dense(80, activation="sigmoid")(x6)
x8 = layers.Dense(80, activation="sigmoid")(x7)
x9 = layers.Dense(80, activation="sigmoid")(x8)
x10= layers.Dense(80, activation="linear")(x9)
outputs = layers.Dense(1)(x10)
model = keras.Model(inputs, outputs)
model.compile(optimizer = 'adam', loss='mean_squared_error')
```

I then began to fit the data to the model to see how well the model geralises the data to similar data within the dataset.
I decided to use 5000 epochs and a batch size of 250 it allowed me to get the lowest loss possible overall.

Definitions

Epochs: "Integer. Number of epochs to train the model. An epoch is an iteration over the entire x and y data provided. Note that in conjunction with initial_epoch, epochs is to be understood as "final epoch". The model is not trained for a number of iterations given by epochs, but merely until the epoch of index epochs is reached."

batch_size: "Integer or None. Number of samples per gradient update. If unspecified, batch_size will default to 32. Do not specify the batch_size if your data is in the form of datasets, generators, or keras.utils.Sequence instances (since they generate batches)."



```python
#https://keras.io/api/models/model_training_apis/
# Fit the data.
model.fit(fulldataset['speed'], fulldataset['power'], epochs=5000, batch_size=250)
```

Plotting of the model and dataset,from what the graph plots the prediction, we can see it is similar to the function but represents the loss as its not as in line with the sigmoid curve 

```python
plt.figure(figsize=(20,10))
plt.title('Wind Turbine Power vs Wind Speed', size = 25)
# Plot the values.
plt.plot(fulldataset['speed'],fulldataset['power'], 'bx')
plt.plot(fulldataset['speed'],model.predict(fulldataset['speed']), 'ro')
plt.xlabel('Speed', size = 25)
plt.ylabel('Power', size = 25)
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.show
```

Saving the model to be used with the webservice

```python
model.save("modelpowerproduction.h5")
```

I wanted to then improve the accuracy of the model and I decided to see if by removing the outliers within the dataset would the loss of the model be improved. 

To do this I decided to to store the dataset fulldataset inside another dataset called datasetNoOut but removed the presence of any of the data with a power of 0. This was decided as after removing the outliers it didnt change the shape of the sigmoid and I deemed it acceptable to remove the outliers.

Please note that by doing this also it removed instances of 0 from the beginning where power should be at 0, so I have not decided to use this as the model for the web service as to keep the it true to the dataset.

```python
# without outliers
# https://machinelearningmastery.com/how-to-use-statistics-to-identify-outliers-in-data/
datasetNoOut = fulldataset[fulldataset.power != 0]
datasetNoOut
```

I then decided to fit the data to the model without the outliers with the same Epoch and Batch Size to get a comparison between the two datasets. with the removal of the outliers , I can see I get a loss of about 16-15

```python
# Fit of the data with outliers removed.
model.fit(datasetNoOut['speed'], datasetNoOut['power'], epochs=5000, batch_size=250)
```

When plotting this dataset to the model we can see the prediction line is nearly identical with the sigmoid curve because with the outliers removed I can get a truer observation of the of the dataset

```python
# Plotting the graph with outliers

plt.figure(figsize=(20,10))
plt.title('Wind Turbine Power vs Wind Speed(With Outliers Removed)', size = 25)
# Plot the values.
plt.plot(datasetNoOut['speed'],datasetNoOut['power'], 'bx')
plt.plot(datasetNoOut['speed'],model.predict(datasetNoOut['speed']), 'ro')
plt.xlabel('Speed', size = 25)
plt.ylabel('Power', size = 25)
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.show
```

## Conclusion

In conclusion after producing both models , one with the outliers inside the data set and the other dataset without the outliers after comparing and contrasting the the two graphs , the graph with the outliers removed creates a graph with the lowest loss and has a prediction lije closest to the sigmoid curve. But keeping in line with the dataset and not changing to keep a truer result representing all of the data I have decided to use that model to use on the web service.
