## Project Overview

For this project I was tasked with producing a model that accurately predicts wind turbine power output from wind speed values from a dataset powerproduction I was given, The idea is to use machine learning to produce an accurate prediction of the dataset.

```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
```

To start of the project, I begin by reading in the dataset from the CSV using pandas,then displaying the dataset to show that the data is being read in successfully. From here we can see that all five hubdred rows of data are successfuly being read in  and it is showing the first five rows and the last five rows

```python
# reads dataset from CSV file given as part of project
dataset = pd.read_csv("https://raw.githubusercontent.com/ianmcloughlin/2020A-machstat-project/master/dataset/powerproduction.csv")
dataset
```

Once the data is successfully read , the next logical step was to visualise to ry and visualise the data to understand the to see how much power was being generated based on the speed of the wind. 

```python
# plotting of data from the dataset to a scatter plot graph
plt.figure(figsize=(20,10))
plt.title('Wind Turbine Power vs Wind Speed', size = 25)
plt.plot(dataset['speed'], dataset['power'],'bx')
plt.xlabel('Speed', size = 25)
plt.ylabel('Power', size = 25)
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.show
```

```python
#with outliers
polyfull = pd.DataFrame(dataset)
polyfull
```

```python
# https://towardsdatascience.com/multi-layer-neural-networks-with-sigmoid-function-deep-learning-for-rookies-2-bf464f09eb7f

inputs = keras.Input(shape=(1,))
# x = layers.experimental.preprocessing.Rescaling(1.0 / 255)(inputs)
# x = layers.Flatten()(x)
x1 = layers.Dense(20, activation="sigmoid")(inputs)
x2 = layers.Dense(80, activation="sigmoid")(x1)
x3 = layers.Dense(80, activation="sigmoid")(x2)
x4 = layers.Dense(80, activation="sigmoid")(x3)
x5 = layers.Dense(80, activation="sigmoid")(x4)
x6 = layers.Dense(80, activation="sigmoid")(x5)
x7 = layers.Dense(80, activation="sigmoid")(x6)
x8 = layers.Dense(80, activation="sigmoid")(x7)
x9 = layers.Dense(80, activation="sigmoid")(x8)
x10= layers.Dense(80, activation="linear")(x9)
outputs = layers.Dense(1)(x10)
model = keras.Model(inputs, outputs)
model.compile(optimizer = 'adam', loss='mean_squared_error')
```

```python
# Fit the data.
model.fit(polyfull['speed'], polyfull['power'], epochs=10000, batch_size=500)
```

```python
plt.figure(figsize=(20,10))
plt.title('Wind Turbine Power vs Wind Speed', size = 25)
# Plot the values.
plt.plot(polyfull['speed'],polyfull['power'], 'bx')
plt.plot(polyfull['speed'],model.predict(polyfull['speed']), 'ro')
plt.xlabel('Speed', size = 25)
plt.ylabel('Power', size = 25)
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.show
```

```python
model.save("modelpowerproduction.h5")
```

```python
# without outliers
# https://machinelearningmastery.com/how-to-use-statistics-to-identify-outliers-in-data/
polyNoOut = polyfull[polyfull.power != 0]
polyNoOut
```

```python
# Fit of the data with outliers removed.
model.fit(polyNoOut['speed'], polyNoOut['power'], epochs=10000, batch_size=500)
```

```python
# Plotting the graph with outliers

plt.figure(figsize=(20,10))
plt.title('Wind Turbine Power vs Wind Speed(With Outliers Removed)', size = 25)
# Plot the values.
plt.plot(polyNoOut['speed'],polyNoOut['power'], 'bx')
plt.plot(polyNoOut['speed'],model.predict(polyNoOut['speed']), 'ro')
plt.xlabel('Speed', size = 25)
plt.ylabel('Power', size = 25)
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.show
```
