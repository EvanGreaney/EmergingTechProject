```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/ianmcloughlin/2020A-machstat-project/master/dataset/powerproduction.csv")

dataset
X= dataset.iloc[:, :-1]
Y= dataset.iloc[:, -1]

X_train, X_test, Y_train, y_test = train_test_split( X, Y, test_size=0.33, random_state=1)
```

```python
plt.figure(figsize=(20,10))
plt.title('Wind Turbine Power vs Wind Speed', size = 25)
plt.plot(dataset['speed'], dataset['power'],'bx')
plt.xlabel('Speed', size = 25)
plt.ylabel('Power', size = 25)
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.show
```

```python

polyfull = pd.DataFrame(dataset)
polyfull
```

```python
polyNoOut = polyfull[polyfull.power != 0]
print(polyNoOut)
```

```python
inputs = keras.Input(shape=(1,))
# x = layers.experimental.preprocessing.Rescaling(1.0 / 255)(inputs)
# x = layers.Flatten()(x)
x1 = layers.Dense(20, activation="sigmoid")(inputs)
x2 = layers.Dense(80, activation="sigmoid")(x1)
x3 = layers.Dense(80, activation="sigmoid")(x2)
x4 = layers.Dense(80, activation="sigmoid")(x3)
x5 = layers.Dense(80, activation="sigmoid")(x4)
x6 = layers.Dense(80, activation="sigmoid")(x5)
x7 = layers.Dense(80, activation="sigmoid")(x6)
x8 = layers.Dense(80, activation="sigmoid")(x7)
x9 = layers.Dense(80, activation="sigmoid")(x8)
x10= layers.Dense(80, activation="linear")(x9)
outputs = layers.Dense(1)(x10)
model = keras.Model(inputs, outputs)
model.summary()
```

```python
# Train a different model.
# model = keras.models.Sequential()
# model.add(layers.LayerNormalization())
# model.add(keras.layers.Dense(50, input_shape=(1,), activation='sigmoid', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))
# model.add(keras.layers.Dense(1, activation='linear', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))
model.compile(optimizer = 'adam', loss='mean_squared_error')
```

```python
# Fit the data.
model.fit(polyfull['speed'], polyfull['power'], epochs=5000, batch_size=1)
```

```python
plt.figure(figsize=(20,10))
plt.title('Wind Turbine Power vs Wind Speed', size = 25)
# Plot the values.
plt.plot(polyfull['speed'],polyfull['power'], 'bx')
plt.plot(polyfull['speed'],model.predict(polyfull['speed']), 'ro')
plt.xlabel('Speed', size = 25)
plt.ylabel('Power', size = 25)
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.show
```
